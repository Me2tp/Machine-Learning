```python
#importing module
import numpy as np
import pandas as pd
```

```python
#fetching Dataset
dataset=pd.read_csv("Startups.csv")
x=dataset.iloc[:,:-1].values
y=dataset.iloc[:, 4].values
```

```python
#categorical encoding
from sklearn.preprocessing import LabelEncoder,OneHotEncoder
from sklearn.compose import ColumnTransformer
l_encoder=LabelEncoder()
x[:,3]=l_encoder.fit_transform(x[:,3])
CT=ColumnTransformer(transformers=[('encode',OneHotEncoder(),[3])],remainder='passthrough')
x=CT.fit_transform(x).astype(np.float)
```

```python
#Avoiding Dummy Trap
x=x[:,1:]
```

```python
#split Test and Train Dataset
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
#fitting Multiple regression in model
from sklearn.linear_model import LinearRegression
regressor=LinearRegression()
regressor.fit(x_train,y_train)
```

```python
#Testing model
y_pred=regressor.predict(x_test)
```

```python
#value of b0
regressor.intercept_
```

```python
#b1x1+b2x2+.........
regressor.coef_
```

```python
#Building Optimum model using Back Elimination 
#this is used to find best fit features
import statsmodels.api as sm
x=np.append(arr=np.ones((50,1)).astype(int),values=x , axis=1) #x0=1
```

```python
x_opt=x[:, [0,1,2,3,4,5]]
regressor_OLS=sm.OLS(endog=y,exog=x_opt).fit()
regressor_OLS.summary()
```

```python
x_opt=x[:, [0,1,3,4,5]]
regressor_OLS=sm.OLS(endog=y,exog=x_opt).fit()
regressor_OLS.summary()
```

```python
x_opt=x[:, [0,3,4,5]]
regressor_OLS=sm.OLS(endog=y,exog=x_opt).fit()
regressor_OLS.summary()
```

```python
x_opt=x[:, [0,3,5]]
regressor_OLS=sm.OLS(endog=y,exog=x_opt).fit()
regressor_OLS.summary()
```

```python
x_opt=x[:, [0,3]]
regressor_OLS=sm.OLS(endog=y,exog=x_opt).fit()
regressor_OLS.summary()
```
